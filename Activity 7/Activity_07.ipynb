```python
from bs4 import BeautifulSoup
import requests
r=requests.get("https://www.geeksforgeeks.org/implementing-web-scraping-python-beautiful-soup/")
print(r)
soup = BeautifulSoup(r.content, 'html.parser')
print(soup.prettify)
print(soup.title)
print(soup.title.name)
print(soup.title.parent.name)
```

```python
s = soup.find('div', class_ ='entry-content')
content = soup.find_all('p')
print(content) 
```

```python
s1 = soup.find('div',class_ ='main')
leftbar = soup.find("ul", class_ = "leftBarList")

content =leftbar.find_all('li')
print(content)
```
